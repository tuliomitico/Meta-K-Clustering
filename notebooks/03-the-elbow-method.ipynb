```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_digits

from src.ranking.vector_measures import get_n_groups_elbow_technique
from src.cluster import KMedoids
```

Instatiating the `MinMaxScaler` model

```python
m, n = 2, 15
mms = MinMaxScaler()
```

Choosing a not preprocessed and meta featured dataset

```python
digits, target = load_digits(return_X_y = True)
scaled_digits = mms.fit_transform(digits)

ng, index = get_n_groups_elbow_technique(scaled_digits,KMedoids,m,n)
```

Plotting the result generated by the function `get_n_groups_elbow_technique`

```python
plt.figure(figsize=(12,6))
plt.plot(range(2,16),index,marker='o',linestyle='--')
plt.xlabel("Number of clusters")
plt.ylabel("Elbow Score")
plt.title("Elbow Technique",fontweight='bold',fontsize=18)
plt.axvline(ng,color='r',linestyle='-')
plt.text(7.25,7250,"Number of cluster optimal is {}".format(ng),color='red',fontsize=16)
```
